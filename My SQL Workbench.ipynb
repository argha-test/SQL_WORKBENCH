- Create Table In MySQL using Python

```python
### Create Department Table 

cursor=connection.cursor()
cursor.execute('''CREATE TABLE department (
              Department_ID INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
              Department_Name VARCHAR(50) NOT NULL UNIQUE,
              Department_Code VARCHAR(20) NOT NULL UNIQUE)''')
```

```python
### Create another table which is employee table
###  also use Foreign key

cursor=connection.cursor()
cursor.execute('''CREATE TABLE employee(
             EmployeeID int NOT NULL AUTO_INCREMENT PRIMARY KEY ,
             Age int,
             Name varchar(50),
             Salary float,
             Address varchar(100),
             email varchar(30) NOT NULL UNIQUE,
             CreatedBy varchar(50),
             CreatedTime DATETIME Default CURRENT_TIMESTAMP,
             ModifiedBy varchar(50),
             ModifiedTime DATETIME,
             Department_Code varchar(20),
             FOREIGN KEY(Department_Code) REFERENCES department(Department_Code))
             ''')
```

- VIEW IN SQL

```python


cursor=connection.cursor()
cursor.execute('''CREATE VIEW 2ND_HIGHEST AS SELECT * FROM employee WHERE Age>30 GROUP BY Salary ORDER BY Salary DESC LIMIT 1,1''' )
```

```python
##Create a SQL view for employees having age>30 and having highest salary.

import pandas as pd
df1=pd.read_sql_query("select * from 2ND_HIGHEST;",connection)
df1
```
